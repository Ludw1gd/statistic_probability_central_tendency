# No. 1

Dalam suatu liga sepak bola, dipilih secara acak 10 pemain dari klub yang berbeda dan diketahui jumlah gol yang mereka raih dalam musim ini adalah: 19, 10, 10, 12, 13, 15, 14, 13, 9, 17. Lakukan analisis terhadap sebaran data dari dataset yang diberikan. Manakah ukuran sebaran yang paling mampu menggambarkan sebaran data dengan baik?

# a. Range

```python
def calculate_range(data):
    nilai_tertinggi = max(data)
    print(f"Nilai tertinggi: {nilai_tertinggi}")

    nilai_terendah = min(data)
    print(f"Nilai terendah: {nilai_terendah}")

    data_range = max(data) - min(data)
    
    print(f"\n")    
    print(f"Range: {nilai_tertinggi} - {nilai_terendah} = {data_range}")  
    
    return data_range


data = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"Data: {data}")

data_range = calculate_range(data)
```

Grafik berikut menggambarkan Rentang (range) secara visual:

```python
import matplotlib.pyplot as plt
import random

hexadecimal = ["#"+''.join([random.choice('ABCDEF0123456789') for i in range(6)])]
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color=hexadecimal)

y = 1.5 
x1 = 4/70*1  
x2 = 66/70*1  

plt.axhline(y=y, xmin=x1, xmax=x2, color='red', linestyle='dashed', linewidth=1, label=f'Range = {calculate_range(data)}')

plt.axvline(x=min(data), color='blue', linestyle='dashed', linewidth=1, label=f'Min Data: {min(data):.2f}')
plt.axvline(x=max(data), color='blue', linestyle='dashed', linewidth=1, label=f'Max Data: {max(data):.2f}')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Range')

plt.legend()

plt.show()
```

# b. Variance

```python
def calculate_mean(data):
    return sum(data) / len(data)

def calculate_variance(data):
    mean = calculate_mean(data)
    print(f"Rata-rata: {mean:.2f}")  
    
    print(f"Kuadrat selisih setiap nilai data dengan rata-rata:")    
    squared_diff = []
    for x in data:
        squared_diff_per_x = (x - mean)**2
        squared_diff.append( squared_diff_per_x )
        print(f"({x} - {mean:.2f})**2 = {(x - mean):.2f}**2 = {squared_diff_per_x:.4f}")    
    
    print(f"Total kuadrat selisih: {sum(squared_diff):.2f}")    
    print(f"Jumlah total data: {len(data)}")    
    
    variance = sum(squared_diff) / len(data)
    
    print(f"\n")    
    print(f"Variansi: {sum(squared_diff):.2f} / {len(data)} = {variance}")    
    
    return variance


data = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"Data: {data}")

variance = calculate_variance(data)
```

# c. Standard Deviation

```python
def calculate_standard_deviation(data):
    variance = calculate_variance(data)
    standard_deviation = variance**0.5
    print(f"\n")   
    print(f"Standard Deviation: {variance}**0.5 = {standard_deviation}")
    
    return standard_deviation

data = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"Data: {data}")

standard_deviation = calculate_standard_deviation(data)
mean = calculate_mean(data)
```

Grafik berikut menunjukkan Standar Deviasi dalam bentuk visual:

```python
import matplotlib.pyplot as plt
import numpy as np
import random

hexadecimal = ["#"+''.join([random.choice('ABCDEF0123456789') for i in range(6)])]

plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color=hexadecimal)

plt.axvline(x=mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean}')
plt.axvline(x=mean - standard_deviation, color='blue', linestyle='dashed', linewidth=1, label=f'{mean-standard_deviation:.2f}')
plt.axvline(x=mean + standard_deviation, color='blue', linestyle='dashed', linewidth=1, label=f'{mean+standard_deviation:.2f}')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title(f'Std Deviation = {standard_deviation:.2f}')

plt.legend()

plt.show()
```

Dari ketiga metode perhitungan di atas, standar deviasi adalah ukuran yang paling mampu menggambarkan sebaran data dengan baik, karena standar deviasi mengukur seberapa jauh data bervariasi dari rata-ratanya dengan menggunakan satuan ukuran yang sama dengan data. Standar deviasi juga lebih sensitif terhadap adanya outlier atau data yang menyimpang jauh dari rata-rata. Semakin besar standar deviasi, semakin besar variasi data, dan sebaliknya.
