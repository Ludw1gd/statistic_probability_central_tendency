# No. 1

Diketahui tinggi badan siswa kelas A adalah 100,5 ; 99,7 ; 101 ; 101,3 ; 99,9 ; 98,9 ; 165 ; 102,2 ; 98,7 ; 102,9 ; dan 100. Menggunakan Mean, Median, dan Modus, coba gambarkan pusat distribusi data berikut. Manakah yang paling bagus?

# a. Mean (Rata-Rata)

```python
data = [100.5 , 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]
print(f"Data: {data}")

jumlah_nilai = sum(data)
print(f"Jumlah semua nilai: {jumlah_nilai: .2f}")

jumlah_total_data = len(data)
print(f"Jumlah total data: {jumlah_total_data}")

mean_value = jumlah_nilai / jumlah_total_data
print(f"Mean: {mean_value: .2f}")
```

Mean dapat digambarkan secara visual sebagai grafik berikut:

```python
import matplotlib.pyplot as plt
import random

hexadecimal = ["#"+''.join([random.choice('ABCDEF0123456789') for i in range(6)])]

plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color=hexadecimal)

plt.axvline(x=mean_value, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean_value:.2f}')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Mean')

plt.legend()

plt.show()
```

# b. Median

```python
data = [100.5 , 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]
print(f"Data: {data}")

sorted_data = sorted(data)
print(f"Data berurut: {sorted_data}")

n = len(sorted_data)
print(f"Jumlah total data: {n}")

index_tengah = n//2 

if n % 2 == 0:
    print(f"Jumlah total data Genap")
    median1 = sorted_data[index_tengah]
    median2 = sorted_data[index_tengah - 1]
    print(f"Nilai tengah: {median1} {median2}")
    median_value = (median1 + median2) / 2
    
else:
    print(f"Jumlah total data Ganjil")
    median_value = sorted_data[index_tengah]
    print(f"Nilai tengah: {median_value}")

print(f"Median: {median_value}")
```

Berikut adalah grafik yang menunjukkan Median dalam bentuk visual:

```python
import matplotlib.pyplot as plt
import random

hexadecimal = ["#"+''.join([random.choice('ABCDEF0123456789') for i in range(6)])]

plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color=hexadecimal)

plt.axvline(x=median_value, color='red', linestyle='dashed', linewidth=1, label=f'Median = {median_value:.2f}')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Median')

plt.legend()

plt.show()
```

# Mode (Modus)

```python
from collections import Counter

data = [100.5 , 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]
print(f"Data: {data}")

counter = Counter(data)
mode_data = dict(counter) 
max_counter = max(list(counter.values())) 
print(f"Max Counter: {max_counter}")

mode_value = []
for k, v in mode_data.items():
    print(f"{k} muncul {v} kali")
    if v == max(list(counter.values())):
        mode_value.append(k)
    
print(f"Mode: {mode_value}")
```

Grafik berikut menunjukkan Modus dalam bentuk visual:

```python
import matplotlib.pyplot as plt
import random

hexadecimal = ["#"+''.join([random.choice('ABCDEF0123456789') for i in range(6)])]

plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color=hexadecimal)

plt.axvline(x=mode_value[0], color='red', linestyle='dashed', linewidth=1, label=f'Modus 1 = {mode_value[0]:.2f}')
plt.axvline(x=mode_value[1], color='blue', linestyle='dashed', linewidth=1, label=f'Modus 2 = {mode_value[1]:.2f}')
plt.axvline(x=mode_value[2], color='red', linestyle='dashed', linewidth=1, label=f'Modus 3 = {mode_value[0]:.2f}')
plt.axvline(x=mode_value[3], color='blue', linestyle='dashed', linewidth=1, label=f'Modus 4 = {mode_value[1]:.2f}')
plt.axvline(x=mode_value[4], color='red', linestyle='dashed', linewidth=1, label=f'Modus 5 = {mode_value[0]:.2f}')
plt.axvline(x=mode_value[5], color='blue', linestyle='dashed', linewidth=1, label=f'Modus 6 = {mode_value[1]:.2f}')
plt.axvline(x=mode_value[6], color='red', linestyle='dashed', linewidth=1, label=f'Modus 7 = {mode_value[0]:.2f}')
plt.axvline(x=mode_value[7], color='blue', linestyle='dashed', linewidth=1, label=f'Modus 8 = {mode_value[1]:.2f}')
plt.axvline(x=mode_value[8], color='red', linestyle='dashed', linewidth=1, label=f'Modus 9 = {mode_value[0]:.2f}')
plt.axvline(x=mode_value[9], color='blue', linestyle='dashed', linewidth=1, label=f'Modus 10 = {mode_value[1]:.2f}')
plt.axvline(x=mode_value[10], color='red', linestyle='dashed', linewidth=1, label=f'Modus 11 = {mode_value[0]:.2f}')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Modes')

plt.legend()

plt.show()
```

Dari hasil perhitungan menggunakan mean, median, dan modus, kita dapat menyimpulkan bahwa median adalah metode yang paling sesuai untuk menghitung nilai rata-rata dari dataset yang diberikan.
